```python
# thêm các thư viện cơ bản để xử lý
import pandas as pd
import mysql.connector
import matplotlib.pyplot as plt

# đọc dữ liệu từ file csv
df = pd.read_csv(
    './cost_revenue_dirty.csv')

# kết nối tới MySQL (port mặc định 3306)
connection = mysql.connector.connect(
    host="localhost",
    user="root",
    password="password",
    database="ex2"
)

# tạo bảng và insert dữ liệu
cursor = mydb.cursor()
cursor.execute("DROP TABLE IF EXISTS cost_revenue_dirty")
cursor.execute("""CREATE TABLE IF NOT EXISTS cost_revenue_dirty (
                id INT AUTO_INCREMENT PRIMARY KEY,
                `rank` INT,
                release_date text,
                movie_title text,
                production_budget text,
                worldwide_gross text,
                domestic_gross text)""")
for index, row in df.iterrows():
    rank = row['Rank']
    release_date = row['Release Date']
    movie_title = row['Movie Title']
    production_budget = row['Production Budget ($)']
    worldwide_gross = row['Worldwide Gross ($)']
    domestic_gross = row['Domestic Gross ($)']
    sql = """INSERT INTO cost_revenue_dirty (`rank`, release_date, movie_title, production_budget, worldwide_gross, domestic_gross)
    VALUES (%s, %s, %s, %s, %s, %s)"""
    val = (rank, release_date, movie_title,
           production_budget, worldwide_gross, domestic_gross)
    mycursor.execute(sql, val)
connection.commit()
# # kiểm tra excute
print(mycursor.rowcount, " records inserted.")
```

```python
import mysql.connector

mydb = mysql.connector.connect(
  host="localhost",
  port="3306",
  user="root",
  password="password",
  database="ex2"
)

mycursor = mydb.cursor()

mycursor.execute("SELECT `rank`, release_date, movie_title, production_budget, worldwide_gross, domestic_gross FROM cost_revenue_dirty")

cost_revenue_dirty = mycursor.fetchall()
cost_revenue_dirty
```

```python
import pandas as pd
import re
cost_revenue_dirty_df = pd.DataFrame(cost_revenue_dirty)
cost_revenue_dirty_df.columns = ['rank', 'release_date', 'movie_title', 'production_budget', 'worldwide_gross', 'domestic_gross']
cost_revenue_dirty_df["worldwide_gross"] = cost_revenue_dirty_df["worldwide_gross"].map(lambda x: re.sub(r'[\$,]', '', x))
cost_revenue_dirty_df["domestic_gross"] = cost_revenue_dirty_df["domestic_gross"].map(lambda x: re.sub(r'[\$,]', '', x))
cost_revenue_dirty_df["worldwide_gross"] = cost_revenue_dirty_df['worldwide_gross'].astype(float)
cost_revenue_dirty_df["domestic_gross"] = cost_revenue_dirty_df['domestic_gross'].astype(float)
cost_revenue_dirty_df["worldwide_gross"] = cost_revenue_dirty_df["worldwide_gross"].map(lambda x: x / 1000000)
head_fifth_movie = cost_revenue_dirty_df.head(5)
head_fifth_movie.dropna(subset=['worldwide_gross', 'production_budget', 'domestic_gross'], inplace=True)
head_fifth_movie = head_fifth_movie.sort_values(by=['worldwide_gross'])
head_fifth_movie
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
f = plt.figure()
# ax.get_yaxis().set_major_formatter(
#     matplotlib.ticker.FuncFormatter(lambda x, p: format(int(x), ',')))
head_fifth_movie.plot(x='movie_title', y='worldwide_gross', kind='bar', ax=f.gca())
plt.ticklabel_format(style = 'plain', axis='y')
plt.xlabel('Movie')
plt.ylabel('Worldwide gross (Milion dollar)')
plt.title('Worldwide gross of fifth head movies')
plt.show()
```
